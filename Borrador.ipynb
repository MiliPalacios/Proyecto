```python
import numpy as np
import pandas as pd
import funciones as f
```

```python
datos = pd.read_excel("datasets/Movimiento.xlsx") 
datos=datos.fillna(int("0"))
for i in range (len(datos.index)) :
        datos["Fecha"][i]=pd.to_datetime(f.texto_a_fechas(datos["Fecha"][i])) 
datos
```

```python
def lotes_sin_nombre(texto):
    try:
        texto=float(texto)
    except:
        texto=float(0)
    return texto
        
```

```python
datos["LOTE"]=datos["LOTE"].apply(lotes_sin_nombre)
datos=datos.sort_values(by=["LOTE"])
lotes=datos["LOTE"].unique()
lotes
```

```python
lote=4
datos_lote=datos[datos["LOTE"].isin([lote])].drop(columns=["LOTE","MES","ESTADO CONTABILIDAD","CUENTA ORIGEN","DESCRIPCION BANCO","OBSERVACIONES","OBSERVACION","Tipo de transaccin","Oficina","Concepto","Documento","Saldo","DETALLE PAGO","BANCO","N. de comprobante","Unnamed: 9"])
datos_lote
```

```python
valores_ingresos=datos_lote["Monto"].sum()
valores_ingresos
```

```python
facturas=datos_lote[datos_lote["FACTURA"]!=0]
valores_facturados=facturas["Monto"].sum()
valores_facturados
```

```python
import matplotlib.pyplot as plt
```

```python
y=[valores_ingresos,valores_facturados]
x=[1,2]
y
```

```python
plt.figure(figsize=(10, 6))
plt.title("CONTROL DEL REGISTRO DE INGRESOS")
barras=plt.bar(x,y,width=0.8,edgecolor="k",color=["g","b"])
plt.ylabel('Cantidad')
plt.xticks([1,2],["INGRESOS","FACTURAS"])
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x()+barra.get_width()/2, altura/2, altura, ha = 'center',va="bottom")
plt.savefig(f'images/i_vs_f_lote_{lote}.png')
plt.show()
```

# REPORTE EN PDF

```python
from fpdf import FPDF
class PDFWithBackground(FPDF):
    def __init__(self):
        super().__init__()
        self.background = None

    def set_background(self, image_path):
        self.background = image_path

    def add_page(self, orientation=''):
        super().add_page(orientation)
        if self.background:
            self.image(self.background, 0, 0, self.w, self.h)

    def footer(self):
        # Posicin a 1.5 cm desde el fondo
        self.set_y(-15)
        # Configurar la fuente para el pie de pgina
        self.set_font('Arial', 'I', 8)
        # Nmero de pgina
        self.cell(0, 10, 'Pgina ' + str(self.page_no()), 0, 0, 'C')
```

```python
pdf=PDFWithBackground()
pdf.set_background("images/background.jpeg")
pdf.add_page()
pdf.set_y(10)
pdf.set_font("Times",size=10)#Arial, Times, Courier
pdf.cell(0,0,"RESUMEN AUTOMTICO",0,1,"L")

pdf.set_y(22)
pdf.set_x(15)
pdf.set_font("Times",style="B",size=13)#Arial, Times, Courier
pdf.cell(0,0,"URBANIZACIN DEL EJRCITO NACIONAL",0,1)

pdf.set_y(40)
pdf.set_x(20)
pdf.set_font("Times",style="B",size=55)#Arial, Times, Courier
pdf.cell(0,0,f"LOTE {lote}",0,1)


pdf.set_y(10)
pdf.set_x(90)
pdf.set_font("Times",style="I",size=10)#Arial, Times, Courier
pdf.cell(0,0,"Elaborado por: Milena Palacios",0,1)

pdf.set_y(275)
pdf.set_font("Times",style="I",size=14)#Arial, Times, Courier
pdf.cell(0,0,"Juntos trabajamos por el bienestar!",0,1,"C")


pdf.output("reportes/Reporte.pdf")
```
